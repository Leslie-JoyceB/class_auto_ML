```python
# Définition de la classe 

class AutoML:
    """
    Classe pour faire du preprocessing et tester des modèles de machine learning
    
    Import des données => read csv
    
    Preprocessing:
    - suppression des NaN (> 20%)
    - remplacement des NaN
    - normalisation 
    
    
    """
    
    def __init__(self, data_filename, target):    
        self.data = pd.read_csv(data_filename)   # load data
        self.data = data_filename
        self.y = self.data[target]               # détermination var dépendante
        self.X = self.data.drop(target, axis =1) # détermination var indépendante(s). Ne pas oublier axis = 1 (col)
        self.scaler = StandardScaler()
    
    def supp_nan (self): 
        # suppression des col contenant que des valeurs uniques et dont le % de NaN > 20%
        for col in self.X.columns:
            if len(np.unique(self.X[col])) == len(self.X): 
                self.X = self.X.drop(col, axis = 1)
            
            elif (self.X[col].isna.sum()* 100)/len(self.X) > 20:
                self.X = self.X.drop(col, axis = 1)
        
        # Remplacer des NaN 
        for col in self.X.columns:
            if col in self.X.select_dtypes(include = [np.number]):
                mean = np.mean(self.X[col])
                self.X[col].fillna(mean, inplace = True)
        
            else:
                self.X[col].fillna(self.X[col].mode, inplace = True)
            
            # for col in self.X.select_dtypes(include = [np.category]): => à tester en lieu et place du else
            # si nécessaire 
        
        ### Def une nouvelle 
        
        # Normalisation
        for col in self.X.columns:
            if col in self.X.select_dtypes(include = [np.number]):
                self.scaler.fit(self.X[col])
                self.scaler.transform(self.X[col])
                    
        
        # Transformation des var catégorielles en var numériques (get dummies) => self.X
        for col in self.X.columns:
            if col in self.X.select_dtypes(include = [np.object]):
                self.X = pd.get_dummies(self.X[col])
        
        # LabelEncoder => y exclusivement car je ne souhaite pas avoir plusieurs cols
            
          
    
    
    
```
